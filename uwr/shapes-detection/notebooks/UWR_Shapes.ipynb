```python
import numpy as np
import onnxruntime as ort
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def load_image(image_path, input_size=640):
    img = Image.open(image_path).convert('RGB')
    img = img.resize((input_size, input_size))
    img = np.array(img).astype(np.float32) / 255.0
    img = np.transpose(img, (2, 0, 1))
    img = np.expand_dims(img, axis=0)
    return img

def draw_predictions(image_path, boxes, scores, labels, class_names):
    image = plt.imread(image_path)
    fig, ax = plt.subplots(1, figsize=(12, 9))
    ax.imshow(image)

    # Initialize a dictionary to count detections for each class
    class_count = {name: 0 for name in class_names}

    for box, score, label in zip(boxes, scores, labels):
        x_min, y_min, x_max, y_max = box
        label_index = int(label)  # Convert to int if not already
        class_name = class_names[label_index]
        class_count[class_name] += 1  # Increment count for the detected class

        # Drawing the bounding box
        rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=2,
                                 edgecolor='red', facecolor='none')
        label_text = f"{class_name} ({score:.2f})"
        ax.add_patch(rect)
        ax.text(x_min, y_min, label_text, verticalalignment='top', color='white', fontsize=12, weight='bold',
                bbox=dict(facecolor='red', alpha=0.5))

    plt.axis('off')
    plt.show()

    # Printing class counts
    print("Detected classes count:")
    for name, count in class_count.items():
        print(f"{name}: {count}")


def postprocess_output(output, conf_threshold=0.5):
    # Extract the tensor from the model output
    predictions = output[0][0]  # Shape (8, 8400)

    # Split the tensor into bounding boxes, objectness scores, and class scores
    boxes = predictions[:4, :]  # First 4 values per anchor are bounding box coords (x_min, y_min, x_max, y_max)
    objectness_scores = predictions[4, :]  # 5th value is the objectness score
    class_scores = predictions[5:, :]  # Remaining values are class probabilities

    # Filter out predictions with low objectness score
    valid_indices = np.where(objectness_scores >= conf_threshold)[0]

    # Select the valid boxes, scores, and class labels
    valid_boxes = boxes[:, valid_indices].T  # Transpose to get (N, 4) format
    valid_scores = objectness_scores[valid_indices]
    valid_labels = class_scores[:, valid_indices].argmax(axis=0)  # Get the class with the highest probability

    return valid_boxes, valid_scores, valid_labels

def main(model_path, image_path, class_names):
    # Load the ONNX model
    ort_session = ort.InferenceSession(model_path)

    # Preprocess the image
    image_tensor = load_image(image_path)

    # Get model's input name
    input_name = ort_session.get_inputs()[0].name

    # Perform inference
    outputs = ort_session.run(None, {input_name: image_tensor})

    # Process the output to get boxes, scores, and labels
    boxes, scores, labels = postprocess_output(outputs, conf_threshold=0.1)

    # Draw bounding boxes on the image
    if len(boxes) > 0:
        draw_predictions(image_path, boxes, scores, labels, class_names)
    else:
        print("No objects detected")

if __name__ == "__main__":
    model_path = '/content/runs/detect/train10/weights/best.onnx'  # Path to the ONNX model
    image_path = '/content/Shapes-1/valid/images/Screenshot-from-2023-10-01-11-38-06_png.rf.a05c175ee8142a8da65805306a7757bb.jpg'  # Path to the image file
    class_names = ["circle", "cross", "square", "triangle"]  # Modify class names based on your model's training

    main(model_path, image_path, class_names)
```

```python
import onnxruntime as ort
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Function to preprocess the input image
def preprocess_image(image_path, input_shape):
    image = cv2.imread(image_path)
    resized_image = cv2.resize(image, (input_shape[2], input_shape[3]))
    rgb_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
    normalized_image = rgb_image / 255.0
    transposed_image = np.transpose(normalized_image, (2, 0, 1)).astype(np.float32)
    input_tensor = np.expand_dims(transposed_image, axis=0)
    return input_tensor

# Function to postprocess the model output and draw boxes on the image
def postprocess_and_draw_boxes(image_path, outputs, conf_threshold=0.1, output_image_path='output_image.jpg'):
    image = cv2.imread(image_path)

    # Iterate over each detection in the output
    for detection in outputs:
        # If detection has a shape of (batch_size, num_detections, detection_attributes), loop over each detection
        if len(detection.shape) == 2:
            for det in detection:
                conf = det[4]  # Confidence score

                if conf >= conf_threshold:
                    x_center, y_center, width, height = det[:4]
                    class_scores = det[5:]
                    class_id = np.argmax(class_scores)
                    class_conf = class_scores[class_id]

                    x1 = int(x_center - width / 2)
                    y1 = int(y_center - height / 2)
                    x2 = int(x_center + width / 2)
                    y2 = int(y_center + height / 2)

                    label = f"Class {class_id}, Conf: {class_conf:.2f}"
                    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        else:
            # In case it's a flat array, handle it similarly
            conf = detection[4]  # Confidence score
            if conf >= conf_threshold:
                x_center, y_center, width, height = detection[:4]
                class_scores = detection[5:]
                class_id = np.argmax(class_scores)
                class_conf = class_scores[class_id]

                x1 = int(x_center - width / 2)
                y1 = int(y_center - height / 2)
                x2 = int(x_center + width / 2)
                y2 = int(y_center + height / 2)

                label = f"Class {class_id}, Conf: {class_conf:.2f}"
                cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imwrite(output_image_path, image)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()

# Function to load and run the ONNX model
def run_onnx_model(model_path, image_path):
    session = ort.InferenceSession(model_path)
    input_name = session.get_inputs()[0].name
    input_shape = session.get_inputs()[0].shape
    input_tensor = preprocess_image(image_path, input_shape)
    outputs = session.run(None, {input_name: input_tensor})[0]
    postprocess_and_draw_boxes(image_path, outputs)

# Test the ONNX model with an image
if __name__ == "__main__":
    model_path = '/content/runs/detect/train9/weights/best.onnx'  # Path to the ONNX model
    image_path = '/content/Shapes-1/valid/images/Screenshot-from-2023-10-01-11-38-06_png.rf.a05c175ee8142a8da65805306a7757bb.jpg'  # Path to the image file
    run_onnx_model(model_path, image_path)
```

```python
import torch
from ultralytics import YOLO
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import imageio

# Load the ONNX model
model = YOLO("/content/runs/best.onnx", task='detect')

# Manually preprocess the frame using Pillow instead of OpenCV
def preprocess_frame(frame, input_size=(640, 640)):
    img = Image.fromarray(frame).convert("RGB")  # Convert frame to RGB
    img = img.resize(input_size)
    img = np.array(img).astype(np.float32)
    img = img / 255.0  # Normalize to [0, 1]
    img = np.transpose(img, (2, 0, 1))  # Convert to CHW format
    img = np.expand_dims(img, axis=0)   # Add batch dimension
    return torch.tensor(img)

# Process video stream frame by frame
def process_video(video_path):
    reader = imageio.get_reader(video_path)  # Open video file or stream

    for frame in reader:
        # Preprocess each frame
        input_tensor = preprocess_frame(frame)

        # Run inference on the frame
        results = model(input_tensor)

        # Display the results
        plt.imshow(results[0].plot())  # results[0] gets the first detection result
        plt.axis('off')  # Hide axes
        plt.pause(0.001)  # Brief pause to allow for frame updating
        plt.clf()  # Clear the previous frame's plot for the next frame

# Example usage
video_path = "/path/to/your/video.mp4"  # Replace with video file path or camera stream URL
process_video(video_path)
```
